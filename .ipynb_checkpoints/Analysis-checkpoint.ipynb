```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
for col in df.columns:
    print(col)
```

```python
df['blueWardsPlacedDiff'] = df['blueWardsPlaced'] - df['redWardsPlaced']
```

```python
df['blueWardsDestroyedDiff'] = df['blueWardsDestroyed'] - df['redWardsDestroyed']
```

```python
df.loc[(df['blueWardsPlacedDiff'] > 0) & (df['blueWins'] == 1)]
```

```python
df.loc[(df['blueWardsPlacedDiff'] < 0) & (df['blueWins'] == 1)]
```

```python
df = df.astype(int)
```

```python
df.loc[0]
```

```python
plt.pie([2113,2440], labels = ['Blue places less wards an wins', 'Blue places more wards and wins'], autopct='%1.1f%%')
```

```python
df.loc[(df['blueWardsPlacedDiff'] > 0) & (df['blueWins'] == 1)]
```

```python
df.loc[(df['blueWardsPlacedDiff'] > 0) & (df['blueWins'] == 0)]
```

```python
plt.pie([2440,2165], labels = ['Blue Places More Wards and Wins', 'Blue Places More Wards and Loses'], autopct='%1.1f%%')
```

```python
moreWardsWin = df.loc[(df['blueWins'] == 1) & (df['blueWardsPlacedDiff'] > 0) | (df['blueWardsPlacedDiff'] < 0) & (df['blueWins'] == 0)].shape[0]
```

```python
moreWardsLose = df.loc[(df['blueWins'] == 1) & (df['blueWardsPlacedDiff'] < 0) | (df['blueWardsPlacedDiff'] > 0) & (df['blueWins'] == 0)].shape[0]
```

```python
plt.pie([moreWardsWin,moreWardsLose], labels = ['Placed More Wards', 'Placed Less Wards'], autopct='%1.1f%%')
```

```python
winnerDestroyedLess = df.loc[(df['blueTowersDestroyed'] < df['redTowersDestroyed']) & (df['blueWins'] == 1) | (df['blueTowersDestroyed'] > df['redTowersDestroyed']) & (df['blueWins'] == 0)].shape[0]
```

```python
winnerDestroyedMore = df.loc[(df['blueTowersDestroyed'] < df['redTowersDestroyed']) & (df['blueWins'] == 0) | (df['blueTowersDestroyed'] > df['redTowersDestroyed']) & (df['blueWins'] == 1)].shape[0]
```

```python
plt.pie([winnerDestroyedMore, winnerDestroyedLess], labels = ['Destroyed More Towers' ,'Destroyed Less Towers'], autopct='%1.1f%%')
```

```python
moreExp = df.loc[(df['blueWins'] == 1) & (df['blueTotalExperience'] > df['redTotalExperience']) | (df['blueWins'] == 0)  & (df['blueTotalExperience'] < df['redTotalExperience'])].shape[0]
```

```python
lessExp = df.loc[(df['blueWins'] == 0) & (df['blueTotalExperience'] > df['redTotalExperience']) | (df['blueWins'] == 1)  & (df['blueTotalExperience'] < df['redTotalExperience'])].shape[0]
```

```python
plt.pie([moreExp,lessExp], labels = ['More Experience Gained', 'Less Experience Gained'], autopct='%1.1f%%')
```

```python
MoreGold = df.loc[(df['blueTotalGold'] > df['redTotalGold']) & (df['blueWins'] == 1) | (df['blueTotalGold'] < df['redTotalGold']) & (df['blueWins'] == 0)].shape[0]
```

```python
LessGold = df.loc[(df['blueTotalGold'] < df['redTotalGold']) & (df['blueWins'] == 1) | (df['blueTotalGold'] > df['redTotalGold']) & (df['blueWins'] == 0)].shape[0]
```

```python
plt.pie([MoreGold,LessGold], labels=['More Gold Earned', 'Less Gold Earned'], autopct='%1.1f%%')
```

```python
FirstBlood = df.loc[(df['blueFirstBlood'] == 1) & (df['blueWins'] == 1) | (df['redFirstBlood'] == 1) & (df['blueWins'] ==0)].shape[0]
NoFirstBlood = df.loc[(df['blueFirstBlood'] == 1) & (df['blueWins'] == 0) | (df['redFirstBlood'] == 1) & (df['blueWins'] ==1)].shape[0]
```

```python
plt.pie([FirstBlood,NoFirstBlood], labels=['Got First Blood', "Didn't Get First Blood"], autopct='%1.1f%%')
```

```python
MoreDragons = df.loc[(df['blueDragons'] > df['redDragons']) & (df['blueWins'] == 1) | (df['blueDragons'] < df['redDragons']) & (df['blueWins'] == 0)].shape[0]
LessDragons = df.loc[(df['blueDragons'] < df['redDragons']) & (df['blueWins'] == 1) | (df['blueDragons'] > df['redDragons']) & (df['blueWins'] ==0)].shape[0]
```

```python
plt.pie([MoreDragons,LessDragons], labels=['More Dragons', 'Less Dragons'], autopct='%1.1f%%')
```

```python
MoreHeralds = df.loc[(df['blueHeralds'] > df['redHeralds']) & (df['blueWins'] == 1) | (df['blueHeralds'] < df['redHeralds']) & (df['blueWins'] == 0)].shape[0]
LessHeralds = df.loc[(df['blueHeralds'] < df['redHeralds']) & (df['blueWins'] == 1) | (df['blueHeralds'] > df['redHeralds']) & (df['blueWins'] ==0)].shape[0]
```

```python
plt.pie([MoreHeralds, LessHeralds], labels=['More Heralds', 'Less Heralds'], autopct='%1.1f%%')
```

```python
MoreKills = df.loc[(df['blueKills'] > df['redKills']) & (df['blueWins'] == 1) | (df['blueKills'] < df['redKills']) & (df['blueWins'] == 0)].shape[0]
LessKills = df.loc[(df['blueKills'] < df['redKills']) & (df['blueWins'] == 1) | (df['blueKills'] > df['redKills']) & (df['blueWins'] ==0)].shape[0]
```

```python
plt.pie([MoreKills, LessKills], labels=['More Kills', 'Less Kills'], autopct='%1.1f%%')
```

```python
MoreAssists = df.loc[(df['blueAssists'] > df['redAssists']) & (df['blueWins'] == 1) | (df['blueAssists'] < df['redAssists']) & (df['blueWins'] == 0)].shape[0]
LessAssists = df.loc[(df['blueAssists'] < df['redAssists']) & (df['blueWins'] == 1) | (df['blueAssists'] > df['redAssists']) & (df['blueWins'] ==0)].shape[0]
```

```python
plt.pie([MoreAssists, LessAssists], labels=['More Assists', 'Less Assists'], autopct='%1.1f%%')
```

```python
MoreJungle = df.loc[(df['blueTotalJungleMinionsKilled'] > df['redTotalJungleMinionsKilled']) & (df['blueWins'] == 1) | (df['blueTotalJungleMinionsKilled'] < df['redTotalJungleMinionsKilled']) & (df['blueWins'] == 0)].shape[0]
LessJungle = df.loc[(df['blueTotalJungleMinionsKilled'] < df['redTotalJungleMinionsKilled']) & (df['blueWins'] == 1) | (df['blueTotalJungleMinionsKilled'] > df['redTotalJungleMinionsKilled']) & (df['blueWins'] ==0)].shape[0]
```

```python
plt.pie([MoreJungle, LessJungle], labels=['More Jungle Minions Killed', 'Less Jungle Minions Killed'], autopct='%1.1f%%')
```

```python
MoreEliteMonsters = df.loc[(df['blueEliteMonsters'] > df['redEliteMonsters']) & (df['blueWins'] == 1) | (df['blueEliteMonsters'] < df['redEliteMonsters']) & (df['blueWins'] == 0)].shape[0]
LessEliteMonsters = df.loc[(df['blueEliteMonsters'] < df['redEliteMonsters']) & (df['blueWins'] == 1) | (df['blueEliteMonsters'] > df['redEliteMonsters']) & (df['blueWins'] ==0)].shape[0]
plt.pie([MoreEliteMonsters, LessEliteMonsters], labels=['More Elite Monsters Killed', 'Less Elite Monsters Killed'], autopct='%1.1f%%')
```

```python
MoreDeaths = df.loc[(df['blueDeaths'] > df['redDeaths']) & (df['blueWins'] == 1) | (df['blueDeaths'] < df['redDeaths']) & (df['blueWins'] == 0)].shape[0]
LessDeaths = df.loc[(df['blueDeaths'] < df['redDeaths']) & (df['blueWins'] == 1) | (df['blueDeaths'] > df['redDeaths']) & (df['blueWins'] ==0)].shape[0]
plt.pie([LessDeaths, MoreDeaths], labels=['Less Deaths', 'More Deaths'], autopct='%1.1f%%')
```
